```python
from statsmodels.tsa.stattools import adfuller, acf, pacf
def check_stationarity(timeseries):
  result = adfuller(timeseries)
  print(f"ADF Statistic {result[0]}")
  print(f"P-Value {result[1]}")
  if result[1] < 0.05:
    print("Data is stationary")
  else:
    print("Data is not stationary")
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
arima_model = SARIMAX(train['Visit_Count'], order = (2, 1, 1))
arima_fit = arima_model.fit(disp=False)
arima_pred = arima_fit.forecast(steps=30)
```

```python
from sklearn.metrics import(
    mean_squared_error as mse,
    mean_absolute_error as mae,
    mean_absolute_percentage_error as mape
)

def performance(actual, predicted):
  print("MAE :", round(mae(actual,predicted), 3))
  print("RMSE :", round(mse(actual,predicted)**0.5, 3))
  print("MAPE :", round(mape(actual, predicted), 3))
```

```python
test
```

```python
plt.figure(figsize=(10,6))
plt.plot(test.index, test['Visit_Count'], label='Actual')
plt.plot(test.index, arima_pred, label='Predicted', linestyle='dashed')
plt.legend()
plt.title('ARIMA Model Forecast')
plt.show()
performance(test['Visit_Count'], arima_pred)
```

```python
sarima_model = SARIMAX(train_df['Visit_Count'], order = (4, 1, 3), seasonal_order=(1, 1, 1, 12), exog = train['Exog'], enforce_invertibility=False, enforce_stationarity=False)
sarima_fit = sarima_model.fit()
sarima_pred = sarima_fit.forecast(steps=30,exog = test['Exog'])
```

```python
plt.figure(figsize=(10,6))
plt.plot(test.index, test['Visit_Count'], label='Actual')
plt.plot(test.index, sarima_pred, label='Predicted', linestyle='dashed')
plt.legend()
plt.title('SARIMAX Model with EXOG Forecast')
plt.show()
performance(test['Visit_Count'], sarima_pred)
```
